Filling the missing vaues for the AQI according to the CPCB + proxy breakpoints for 12 pollutants

```python


# CPCB + proxy breakpoints for 12 pollutants
aqi_breakpoints = {
    'PM2.5': [(0, 30, 0, 50), (31, 60, 51, 100), (61, 90, 101, 200), (91, 120, 201, 300), (121, 250, 301, 400), (251, 350, 401, 500)],
    'PM10': [(0, 50, 0, 50), (51, 100, 51, 100), (101, 250, 101, 200), (251, 350, 201, 300), (351, 430, 301, 400), (431, 500, 401, 500)],
    'NO2': [(0, 40, 0, 50), (41, 80, 51, 100), (81, 180, 101, 200), (181, 280, 201, 300), (281, 400, 301, 400), (401, 1000, 401, 500)],
    'SO2': [(0, 40, 0, 50), (41, 80, 51, 100), (81, 380, 101, 200), (381, 800, 201, 300), (801, 1600, 301, 400), (1601, 2000, 401, 500)],
    'CO': [(0.0, 1.0, 0, 50), (1.1, 2.0, 51, 100), (2.1, 10.0, 101, 200), (10.1, 17.0, 201, 300), (17.1, 34.0, 301, 400), (34.1, 50.0, 401, 500)],
    'O3': [(0, 50, 0, 50), (51, 100, 51, 100), (101, 168, 101, 200), (169, 208, 201, 300), (209, 748, 301, 400), (749, 1000, 401, 500)],
    'NH3': [(0, 200, 0, 50), (201, 400, 51, 100), (401, 800, 101, 200), (801, 1200, 201, 300), (1201, 1800, 301, 400), (1801, 2000, 401, 500)],
    'NO': [(0, 50, 0, 50), (51, 100, 51, 100), (101, 200, 101, 200), (201, 400, 201, 300), (401, 800, 301, 400), (801, 1600, 401, 500)],
    'NOx': [(0, 100, 0, 50), (101, 200, 51, 100), (201, 400, 101, 200), (401, 800, 201, 300), (801, 1600, 301, 400), (1601, 2000, 401, 500)],
    'Benzene': [(0, 5, 0, 50), (6, 10, 51, 100), (11, 15, 101, 200), (16, 25, 201, 300), (26, 50, 301, 400), (51, 100, 401, 500)],
    'Toluene': [(0, 100, 0, 50), (101, 200, 51, 100), (201, 400, 101, 200), (401, 800, 201, 300), (801, 1600, 301, 400), (1601, 2000, 401, 500)],
    'Xylene': [(0, 100, 0, 50), (101, 200, 51, 100), (201, 400, 101, 200), (401, 800, 201, 300), (801, 1600, 301, 400), (1601, 2000, 401, 500)],
}

# Function to calculate AQI for a single pollutant
def calc_individual_aqi(conc, pollutant):
    if pd.isnull(conc) or pollutant not in aqi_breakpoints:
        return None
    for bp_lo, bp_hi, i_lo, i_hi in aqi_breakpoints[pollutant]:
        if bp_lo <= conc <= bp_hi:
            return round(((i_hi - i_lo) / (bp_hi - bp_lo)) * (conc - bp_lo) + i_lo)
    return None

# Function to compute overall AQI for a row
def compute_aqi(row):
    pollutants = list(aqi_breakpoints.keys())  # Use all 12
    aqi_vals = [calc_individual_aqi(row.get(p), p) for p in pollutants if pd.notnull(row.get(p))]
    aqi_vals = [val for val in aqi_vals if val is not None]
    return max(aqi_vals) if aqi_vals else np.nan

# Calculate AQI only for missing values and fill them
data['AQI'] = data.apply(lambda row: compute_aqi(row) if pd.isnull(row['AQI']) else row['AQI'], axis=1)
```

Filling the AQI_Bucket for the new generated values for the missing values without changing the original dataset

```python
def classify_aqi(aqi):
    if pd.isna(aqi):
        return np.nan
    elif aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Satisfactory'
    elif aqi <= 200:
        return 'Moderate'
    elif aqi <= 300:
        return 'Poor'
    elif aqi <= 400:
        return 'Very Poor'
    elif aqi <= 500:
        return 'Severe'
    else:
        return 'Hazardous'

data['AQI_Bucket'] = data['AQI'].apply(classify_aqi)
```

# OBJECTIVE-3 : Yearly AQI Trend Analysis - Analyzing Changes in AQI Over Time

```python
# Ensure datetime format is correct
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)

# Extract year and making new column for it
data['Year'] = data['Date'].dt.year

# Group by year and calculate mean AQI
yearly_avg = data.groupby('Year', as_index=False)['AQI'].mean()

# Increase figure size
plt.figure(figsize=(10, 6))

# Plot AQI trend year-wise
sns.lineplot(data=yearly_avg, x='Year', y='AQI', marker='o')
```
